```python
# importando a biblioteca Pandas
import pandas as pd
```

```python
# lendo um arquivo CSV
df = pd.read_csv('house-prices.csv')

```

```python
df
```

```python
# calculando a média de uma coluna
media = df['Price'].mean()

# calculando a mediana de uma coluna
mediana = df['Price'].median()

# calculando a moda de uma coluna
moda = df['Neighborhood'].mode()[0]

# calculando a soma de uma coluna
soma = df['Price'].sum()

#calculando o desvio padrão de uma coluna
std = df['Price'].std()
```

```python
print(f'''
 Valor médio = {media}
 Valor mediano = {mediana}
 Região com mais oferta de imóveis = {moda}
 Soma = {soma}
 Desvio Padrão = {std}


''')
```
