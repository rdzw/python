```python
# importando a biblioteca Pandas
import pandas as pd
```

```python
# Lendo arquivos HTML
df = pd.read_html('https://gist.github.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6')[0]
```

```python
df.isnull().sum()
```

```python
# renomeando colunas
df.rename(columns={'Attack': 'Ataque'}, inplace=True)

# removendo colunas
df.drop(columns=['Unnamed: 0', '#'], inplace=True)
```

```python
df
```

```python
# agrupando dados
df.groupby('Type 1').mean()
```

```python
#Mais flexivel ainda, podemos escolher o tipo de operaçao por coluna
df.groupby('Type 1').aggregate({"Ataque" : "mean"
                                   , "Defense" : "median"
                                   , "Speed" : "max"})
```
