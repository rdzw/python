```python
import pandas as pd
```

```python
df = pd.read_excel('dados_niveis.xlsx')
```

```python
df
```

```python
# verificando a quantidade de dados faltantes em cada coluna
df.isnull().sum()
```

```python
df['Nível (m)'].plot()
```

```python
media = df['Nível (m)'].mean()
```

```python
# preenchendo dados faltantes com um valor específico
df.fillna(media, inplace=True)
```

```python
df
```

```python
#Ler os dados novamente
df = pd.read_excel('dados_niveis.xlsx')
```

```python
# removendo linhas com dados faltantes
df.dropna(inplace=True)
```

```python
df['Nível (m)']
```

```python
#Ler novamente
df = pd.read_excel('dados_niveis.xlsx')

# Por fim, um método mais sofisticado é estimando uma regressao dos dados faltantes para estimar os faltantes
df['Nível (m)'] = df['Nível (m)'].interpolate(method ='linear', limit_direction ='forward') 
# Este é um dos métodos mais utilizados para o tratamento de missing values
```

```python
df['Nível (m)'].plot()
```
